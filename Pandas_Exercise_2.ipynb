6. > Add a new column which indicates if the sum of the rows is Above 90

We will use a for loop

```python
my_mat=np.random.randint(0,20,(10,10))  # generating the matrix from fresh
my_df=pd.DataFrame(my_mat, columns='A B C D E F G H I J'.split())
```

```python
my_df['Total']=0 # First create a new column and the associate values to 0
for i in range(10):
    my_df['Total'][i]=sum(my_df.iloc[i,:].values)   # Run a for loop and fill new new columns with sum of the rows
```

```python
my_df
```

```python
my_df['Total']=['HIGH' if x >=90 else 'LOW' for x in my_df['Total']]
```

```python
my_df
```
